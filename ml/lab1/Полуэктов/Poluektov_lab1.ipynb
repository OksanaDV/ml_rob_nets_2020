9. Вычислите корреляцию Пирсона и Спирмена между признаками Gender и Missions. Сравните полученные величины корреляции. Постройте гистограммы для указанных признаков. Оцените значение p-value.

```python
from scipy import stats
from sklearn.preprocessing import LabelEncoder
from scipy.stats import pearsonr
from scipy.stats import spearmanr
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
laber_encoder = LabelEncoder()
Gen = pd.Series(laber_encoder.fit_transform(cosmos['Gender'].astype(str)))
Mis = pd.Series(laber_encoder.fit_transform(cosmos['Missions'].astype(str)))

korPears = pearsonr(Gen, Mis )
korSpearm = spearmanr(Gen, Mis)

print("Значение корреляции Пирсона: ", korPears)
print("Значение корреляции Спирмена: ", korSpearm)

if (korSpearm[1]<0.05):
    print("p-value is good")
else:
    print("p-value is not good")
```
